# **Step 9: Perform DPO/ORPO Preference Alignment**

Use your preference dataset and perform DPO alignment.

The goal is to improve the model's response quality after instruction fine-tuning.

Load Stage 2 Model

```python
print("Loading Stage 2 model...")

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name=HF_REPOS["stage2"],
    max_seq_length=MAX_SEQ_LENGTH,
    load_in_4bit=True,
)

print("✅ Stage 2 Model Loaded")
```

Attach DPO LoRA

```python
model = FastLanguageModel.get_peft_model(
    model,
    r=8,
    target_modules=[
        "q_proj",
        "k_proj",
        "v_proj",
        "o_proj",
        "gate_proj",
        "up_proj",
        "down_proj",
    ],
    lora_alpha=16,
    lora_dropout=0,
    bias="none",
    use_gradient_checkpointing="unsloth",
)

print("✅ DPO LoRA Attached")
```

Format Dataset (Alpaca)

```python
{
  "prompt":"...",
  "chosen":"...",
  "rejected":"..."
}
```

Load Preference Dataset

```python
from datasets import load_dataset

# Load only the specific .jsonl file directly using your HF Repo ID
dataset = load_dataset(
    HF_REPOS["dataset"],
    data_files="insurance_preference_dataset.jsonl"
)

# Access your data inside the 'train' split
print(f"✅ Successfully loaded specific file!")
print(f"Total records: {len(dataset['train'])}")
```

Fomrat Dataset

```python
EOS_TOKEN = tokenizer.eos_token

def format_dpo_dataset(examples):

    prompts = [
        f"""Below is an instruction that describes a task.
Write a response that appropriately completes the request.

### Instruction:
{prompt}

### Response:
""" + EOS_TOKEN
        for prompt in examples["prompt"]
    ]

    return {
        "prompt": prompts,
        "chosen": examples["chosen"],
        "rejected": examples["rejected"],
    }

dataset = dataset.map(
    format_dpo_dataset,
    batched=True,
)

print("✅ Alpaca DPO Formatting Applied")
```

Validate Dataset

```python
print("PROMPT")
print(dataset["train"][0]["prompt"])

print("\nCHOSEN")
print(dataset["train"][0]["chosen"])

print("\nREJECTED")
print(dataset["train"][0]["rejected"])
```

Configure DPO

```python
dpo_config = DPOConfig(
    max_length=MAX_SEQ_LENGTH,
    max_prompt_length=1024,
    per_device_train_batch_size=1,
    gradient_accumulation_steps=4,
    max_steps=50,
    warmup_steps=5,
    learning_rate=5e-6,
    beta=0.1,
    fp16=True,
    bf16=False,
    logging_steps=5,
    output_dir="outputs_dpo",
    report_to="none",
    save_strategy="no",
)
```

Create DPO Trainer

```python
trainer = DPOTrainer(
    model=model,
    ref_model=None,
    args=dpo_config,
    train_dataset=dataset["train"],
    processing_class=tokenizer,
)

print("✅ DPO Trainer Ready")
```

 Train

```python
trainer.train()
print("✅ DPO Training Complete")
```

Alpaca Inference Function

```python
def generate_answer(question):

    prompt = f"""Below is an instruction that describes a task.
Write a response that appropriately completes the request.

### Instruction:
{question}

### Response:
"""

    inputs = tokenizer(
        prompt,
        return_tensors="pt",
    ).to("cuda")

    with torch.no_grad():

        outputs = model.generate(
            **inputs,
            max_new_tokens=250,
            do_sample=False,
            temperature=0.1,
            use_cache=True,
            eos_token_id=tokenizer.eos_token_id,
        )

    text = tokenizer.decode(
        outputs[0],
        skip_special_tokens=True,
    )

    answer = text.split(
        "### Response:"
    )[-1].strip()

    return answer
```

```python
# =====================================================
# Switch To Inference Mode
# =====================================================

FastLanguageModel.for_inference(model)

model.eval()

torch.set_grad_enabled(False)

gc.collect()
torch.cuda.empty_cache()

print("✅ Inference mode enabled")
```

Build Stage 3 Evaluation Report

```python
# =====================================================
# Download Latest Stage 2 Evaluation Report
# =====================================================

import os
from datetime import datetime

from huggingface_hub import (
    HfApi,
    hf_hub_download,
)

api = HfApi(token=HF_TOKEN)

INPUT_REPO_ID = HF_REPOS["stage2"]
OUTPUT_REPO_ID = HF_REPOS["stage3"]

# =====================================================
# Find Latest Stage 2 Report
# =====================================================

files = api.list_repo_files(
    repo_id=INPUT_REPO_ID,
    repo_type="model",
)

report_files = [
    f
    for f in files
    if (
        f.startswith("results/")
        and "stage2_evaluation_report_" in f
        and f.endswith(".md")
    )
]

if not report_files:
    raise RuntimeError(
        "No Stage 2 evaluation reports found."
    )

latest_report = sorted(report_files)[-1]

print(
    f"✅ Latest Stage 2 Report:\n"
    f"{latest_report}"
)

# =====================================================
# Download Report
# =====================================================

local_stage2_report = hf_hub_download(
    repo_id=INPUT_REPO_ID,
    filename=latest_report,
    repo_type="model",
)

print(
    f"✅ Downloaded:\n"
    f"{local_stage2_report}"
)

# =====================================================
# Read Report
# =====================================================

with open(
    local_stage2_report,
    "r",
    encoding="utf-8",
) as f:

    lines = f.readlines()

# =====================================================
# Parse Existing Table
# =====================================================

records = []

for line in lines:

    line = line.strip()

    if not line.startswith("|"):
        continue

    if (
        "Question" in line
        or "---" in line
        or ":---" in line
    ):
        continue

    cols = [
        c.strip()
        for c in line.split("|")[1:-1]
    ]

    if len(cols) < 5:
        continue

    records.append(
        (
            cols[0],  # Question No.
            cols[1],  # Question
            cols[2],  # Base Model Answer
            cols[3],  # Non-Instruction Model Answer
            cols[4],  # Instruction Model Answer
        )
    )

print(
    f"✅ Parsed {len(records)} questions"
)

if not records:
    raise RuntimeError(
        "No valid evaluation records found inside Stage 2 report."
    )

# =====================================================
# Generate Stage 3 Evaluation Report
# =====================================================

print(
    f"\n🚀 Generating Stage 3 Evaluation Report "
    f"({len(records)} questions)..."
)

markdown = [
    "# Stage 3 Evaluation Report",
    "",
    "| Question No. | Question | Base Model Answer | Non-Instruction Model Answer | Instruction Model Answer | Preference Model Answer |",
    "|---|---|---|---|---|---|",
]

for idx, (
    question_no,
    question,
    base_answer,
    stage1_answer,
    stage2_answer,
) in enumerate(records, start=1):

    print(
        f"[{idx}/{len(records)}] Evaluating..."
    )

    clean_question = (
        question
        .replace("<br>", "\n")
        .replace("<br/>", "\n")
    )

    stage3_answer = generate_answer(
        clean_question
    )

    safe_question = (
        question
        .replace("|", "\\|")
    )

    safe_base_answer = (
        base_answer
        .replace("|", "\\|")
    )

    safe_stage1_answer = (
        stage1_answer
        .replace("|", "\\|")
    )

    safe_stage2_answer = (
        stage2_answer
        .replace("|", "\\|")
    )

    safe_stage3_answer = (
        stage3_answer
        .replace("\n", "<br>")
        .replace("|", "\\|")
    )

    markdown.append(
        f"| {question_no} "
        f"| {safe_question} "
        f"| {safe_base_answer} "
        f"| {safe_stage1_answer} "
        f"| {safe_stage2_answer} "
        f"| {safe_stage3_answer} |"
    )

print(
    "✅ Stage 3 Evaluation Complete"
)

# =====================================================
# Save Report
# =====================================================

timestamp = datetime.now().strftime(
    "%Y%m%d_%H%M%S"
)

output_report_path = (
    f"/content/reports/"
    f"stage3_evaluation_report_{timestamp}.md"
)

os.makedirs(
    "/content/reports",
    exist_ok=True,
)

with open(
    output_report_path,
    "w",
    encoding="utf-8",
) as f:

    f.write(
        "\n".join(markdown)
    )

print(
    "\n✅ Stage 3 Report Saved"
)

print(
    output_report_path
)
```
