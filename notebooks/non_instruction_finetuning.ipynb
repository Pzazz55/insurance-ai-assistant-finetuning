Extract PDF Text

```python
def extract_pdf_text(pdf_path):
    doc = fitz.open(pdf_path)
    text = []
    for page in doc:
        text.append(page.get_text())

    return "\n".join(text)


from huggingface_hub import hf_hub_download

pdf_path = hf_hub_download(
    repo_id=HF_REPOS["dataset"],
    filename="insurance_rawdata.pdf",
    repo_type="dataset"
)


raw_text = extract_pdf_text(pdf_path)

from datasets import Dataset

chunk_size = 1000

chunks = []

for i in range(
    0,
    len(raw_text),
    chunk_size,
):

    chunk = raw_text[
        i : i + chunk_size
    ].strip()

    if len(chunk) > 50:

        chunks.append(chunk)

raw_dataset = Dataset.from_dict(
    {
        "text": chunks
    }
)

print(
    f"✅ Created {len(chunks)} text chunks"
)

print("✅ PDF Extracted")
```

# **Step 5: Test the Base Model**

Before instruction fine-tuning, test the original base model on 10 domain-specific questions.
The goal is to understand how the base model performs before training.

Evaluate Base Model

```python
# =====================================================
# Base Model Inference Function
# =====================================================

def generate_answer(question):

    prompt = f"""
Below is an instruction that describes a task.

Write a response that appropriately completes the request.

### Instruction:
{question}

### Response:
"""

    inputs = tokenizer(
        prompt,
        return_tensors="pt",
    ).to("cuda")

    with torch.no_grad():

        outputs = model.generate(
            **inputs,
            max_new_tokens=250,
            do_sample=False,
            temperature=0.1,
            use_cache=True,
            eos_token_id=tokenizer.eos_token_id,
        )

    text = tokenizer.decode(
        outputs[0],
        skip_special_tokens=True,
    )

    answer = text.split(
        "### Response:"
    )[-1].strip()

    return answer

# =====================================================
# Evaluate Base Model
# =====================================================

base_model_answers = []

print(
    "\n🚀 Evaluating Base Model..."
)

for idx, question in enumerate(
    evaluation_questions,
    start=1,
):

    print(
        f"[{idx}/{len(evaluation_questions)}] "
        f"Base Model..."
    )

    answer = generate_answer(
        question
    )

    base_model_answers.append(
        answer
    )

print(
    "✅ Base Model Evaluation Complete"
)
```

Attach LoRA

```python
model = FastLanguageModel.get_peft_model(
    model,
    r=16,
    target_modules=[
        "q_proj",
        "k_proj",
        "v_proj",
        "o_proj",
        "gate_proj",
        "up_proj",
        "down_proj",
    ],
    lora_alpha=32,
    lora_dropout=0,
    bias="none",
    use_gradient_checkpointing="unsloth",
)

print("✅ LoRA Attached")
```

Tokenize

```python
# =====================================================
# Tokenize Dataset
# =====================================================

def tokenize_fn(examples):

    return tokenizer(
        examples["text"],
        truncation=True,
        max_length=MAX_SEQ_LENGTH,
    )


tokenized_dataset = raw_dataset.map(
    tokenize_fn,
    batched=True,
    remove_columns=["text"],
)

print(
    f"✅ Tokenized Records: "
    f"{len(tokenized_dataset)}"
)

print(
    tokenized_dataset.features
)
```

Train Stage 1

```python
from transformers import (
    Trainer,
    TrainingArguments,
    DataCollatorForLanguageModeling,
)

# =====================================================
# Filter Very Short Examples
# =====================================================

print(
    f"Before filtering: {len(tokenized_dataset)}"
)

tokenized_dataset = tokenized_dataset.filter(
    lambda x: len(x["input_ids"]) > 10
)

print(
    f"After filtering: {len(tokenized_dataset)}"
)

# =====================================================
# Dataset Statistics
# =====================================================

lengths = [
    len(x["input_ids"])
    for x in tokenized_dataset
]

print(
    f"Min length : {min(lengths)}"
)

print(
    f"Max length : {max(lengths)}"
)

print(
    f"Avg length : "
    f"{sum(lengths)/len(lengths):.2f}"
)

# =====================================================
# Stage 1 Training
# =====================================================

trainer = Trainer(
    model=model,
    train_dataset=tokenized_dataset,
    data_collator=DataCollatorForLanguageModeling(
        tokenizer=tokenizer,
        mlm=False,
    ),
    args=TrainingArguments(
        per_device_train_batch_size=1,
        gradient_accumulation_steps=4,
        warmup_steps=5,
        max_steps=60,
        learning_rate=2e-4,
        fp16=True,
        logging_steps=5,
        output_dir="outputs_raw",
        save_strategy="no",
        report_to="none",
    ),
)

# =====================================================
# Start Training
# =====================================================

print(
    "\n🚀 Starting Stage 1 Training..."
)

trainer.train()

print(
    "\n✅ Stage 1 Training Complete"
)
```
